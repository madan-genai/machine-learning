# 1. Loading the Image

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("/content/image.jpg")
```

# 2. Converting Image Color Formats

```python
img_rbg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_rbg)
plt.show()
```

# 3. Loading Haar Cascade Classifier

```python
stop_cascade=cv2.CascadeClassifier("/content/stop_data.xml")
```

# 4. Detecting Objects in the Image

```python
found=stop_cascade.detectMultiScale(img_gray,minSize=(20,20))
```

```python
found
```

# 5. Drawing Rectangles Around Detected Objects

```python
for (x,y,w,h) in found:
  cv2.rectangle(img_rbg,(x,y),(x+w,y+h),(0,255,0),5)
```

# 6. Displaying the Result

```python
plt.imshow(img_rbg)
plt.show()
```
