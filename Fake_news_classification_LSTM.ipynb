```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

# ONE HOT REPRESTATION

```python
messages=X.copy()
```

```python
messages["article_title"][1]
```

```python
messages
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download("stopwords")
```

```python
## data preprocessing
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0, len(messages)):
  review = re.sub('[^a-zA-Z]', ' ', messages['article_title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
  review = " ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_rep=[one_hot(words,voc_size) for words in corpus]
onehot_rep[1]
```

# Embedding Represtation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_rep,padding="pre",maxlen=sent_length)
print(embedded_docs)
```

```python
embedding_vector_features = 40

model = Sequential()

model.add(Embedding(input_dim=voc_size,
                    output_dim=embedding_vector_features,
                    input_shape=(sent_length,)))

model.add(LSTM(100))
model.add(Dense(1, activation="sigmoid"))

model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])

model.summary()
```

```python
len(embedded_docs), Y.shape
```

```python
import numpy as np
```

```python
X_final=np.array(embedded_docs)
Y_final=np.array(Y)
```

```python
X_final.shape,Y_final.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_final,Y_final,test_size=0.33,random_state=42)
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=10,batch_size=64)
```

# Adding Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
#creating Model
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(input_dim=voc_size,
                    output_dim=embedding_vector_features,
                    input_shape=(sent_length,)))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
model.summary()
```

# Performance Metrics And Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
